```python
import os
import pandas as pd

def process_csv_files(root_dirs):
    # 遍历所有顶级文件夹
    for root_dir in root_dirs:
        print(f"Checking directory: {root_dir}")
        if os.path.exists(root_dir):
            print(f"Processing directory: {root_dir}")
        else:
            print(f"Directory not found: {root_dir}")
            continue  # 直接跳过不存在的目录
        
        # 确定子文件夹 'bfs' 和 'randoms'
        subfolders = ['bfs', 'randoms']
        for subfolder in subfolders:
            subfolder_path = os.path.join(root_dir, subfolder)
            if os.path.isdir(subfolder_path):
                print(f"Processing subdirectory: {subfolder_path}")
            else:
                print(f"Subdirectory not found: {subfolder_path}")
                continue
            
            # 处理子文件夹中的每个 CSV 文件
            csv_files_found = False
            for file_name in os.listdir(subfolder_path):
                if file_name.endswith('.csv'):
                    csv_files_found = True
                    file_path = os.path.join(subfolder_path, file_name)
                    try:
                        # 读取 CSV 文件并计算列的总和
                        match_data = pd.read_csv(file_path)
                        individual_sums = match_data[['eye', 'emo', 'mouth']].sum()
                        print(f"Sums for {file_path}:")
                        print(individual_sums)
                    except Exception as e:
                        print(f"Error processing file {file_path}: {e}")
            if not csv_files_found:
                print(f"No CSV files found in {subfolder_path}")

# 列出所有顶级文件夹
top_folders = ['C62', 'C63', 'C64', 'C65','C1andC66']
# 给出顶级文件夹的全路径，假设它们位于当前脚本的相同目录下
top_folder_paths = [os.path.join(os.getcwd(), folder) for folder in top_folders]

process_csv_files(top_folder_paths)
```

```python
import os
import pandas as pd

def find_max_sum_csv(root_dirs):
    max_sum = 0
    max_file_path = None
    
    # 遍历所有顶级文件夹
    for root_dir in root_dirs:
        # 确定子文件夹 'bfs' 和 'randoms'
        subfolders = ['bfs', 'randoms']
        for subfolder in subfolders:
            subfolder_path = os.path.join(root_dir, subfolder)
            if os.path.isdir(subfolder_path):
                # 处理子文件夹中的每个 CSV 文件
                for file_name in os.listdir(subfolder_path):
                    if file_name.endswith('.csv'):
                        file_path = os.path.join(subfolder_path, file_name)
                        try:
                            # 读取 CSV 文件并计算列的总和
                            match_data = pd.read_csv(file_path)
                            total_sum = match_data[['eye', 'emo', 'mouth']].sum().sum()
                            if total_sum > max_sum:
                                max_sum = total_sum
                                max_file_path = file_path
                        except Exception as e:
                            print(f"Error processing file {file_path}: {e}")

    # 打印总和最大的文件路径和总和
    if max_file_path:
        print(f"The file with the highest sum of 'eye', 'emo', 'mouth' is {max_file_path} with a sum of {max_sum}")

# 列出所有顶级文件夹
top_folders = ['C62', 'C63', 'C64', 'C65','C1andC66']
top_folder_paths = [os.path.join(os.getcwd(), folder) for folder in top_folders]

find_max_sum_csv(top_folder_paths)
```
